```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
F = 10  # N
delta = 10 / 1000  # m
k = F / delta  # N/m
G = 69E9  # Pa

# Values of D to loop through
# Generate wire diameter array
diameter1 = np.array([12, 18, 25]) # spring diameter in mm
diameter1_m = diameter1 / 1000  # convert to m
diameter = diameter1.tolist()  # convert to list
Dia = diameter1_m.tolist()

#D_values = [12 / 1000, 18 / 1000, 25 / 1000] #wire diameters

# Create a figure and axis objects for plotting
fig, ax = plt.subplots()


# Generate wire diameter array
wire_d1 = np.linspace(0.1, 2, 100)  # wire diameter in mm
wire_d1_m = wire_d1 / 1000  # convert to m
wire_d = wire_d1_m.tolist()  # convert to list


# Loop through different values of D
i = 0
for D in Dia:

    # Calculate the number of turns for each wire diameter
    Na = [d**4 * G / (8 * D**3 * k) for d in wire_d]

    # Plot wire diameter vs. number of turns
    ax.plot(wire_d1, Na, label=f"D = {diameter[i]} mm")
    
    i = i + 1

# Set labels and title
ax.set_xlabel('Wire Diameter (m)')
ax.set_ylabel('Number of Turns')
ax.set_title('Number of Turns vs. Wire Diameter')

# Add legend
ax.legend()

# Show plot
plt.show()
```
